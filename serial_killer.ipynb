```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Identified Serial killers in the United States in 60 years (1960-2020) - Killer's Information.csv")
df1 = pd.read_csv("Identified Serial killers in the United States in 60 years (1960-2020) - Study Case.csv")
```

```python
Sex = df["Sex"].value_counts()

s = ["Male", "Female"]

plt.title("Sex of the Serial Killers",fontsize=14, fontweight="bold")
plt.pie(Sex, labels=s, shadow=True, explode=(0,0.5), colors=("#5584AC","#FFA8A8"), autopct = '%1.1f%%')
plt.show()
```

```python
Age = df["Age Group (While committed crime)"].value_counts()

a = ["Adult","Minor","Middle Age Adult","Senior Adult"]

plt.title("Age Group of the Serial Killers",fontsize=14, fontweight="bold")
plt.pie(Age, labels=a, shadow=True, explode=(0,0.3,0.2,0.1), colors=("#B983FF","#96C7C1","#E79E85","#FCF2A7"), autopct = '%1.1f%%')
plt.show()
```

```python
Sex = df["Sex"].value_counts()

s = ["Male", "Female"]

plt.title("Sex of the Serial Killers",fontsize=14, fontweight="bold")
plt.pie(Sex, labels=s, shadow=True, explode=(0,0.5), colors=("#5584AC","#FFA8A8"), autopct = '%1.1f%%')
plt.show()
```

```python
Status = df["Status"].value_counts()

st = ["Sentenced to life imprisonment","Executed","Died in prison","Sentenced to death","Committed Suicide","Incarcerated","Killed","Released","Deceased","Died in Health Facility","Awaiting sentencing","Escaped","Unknown"]

plt.figure(figsize=(9,8))
plt.title("Status of the Serial Killers",fontsize=16)
plt.barh(st, Status, color=("#852747","#FF8474","#0093AB","#DA723C","#6BCB77","#FFB72B","#655D8A","#FD2E2E","#5E616A","#B3541E","#781D42"))
plt.show()
```

```python
ys = []
lp = []
for i in range(1960,2020):
  ys.append(i)
for j in ys:
  lp.append(len(df[df["Years Active (Start)"] == j]))

plt.figure(figsize=(15,5))
plt.title("The Numbers of the Killers that Started to Active In Each Year", fontsize=14)
plt.xlabel("Years", fontsize=12)
plt.plot(ys, lp, color = "#4C3F91", linewidth=3)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import folium

m = folium.Map(location=[40, -95], zoom_start=4)
url = (
    "https://raw.githubusercontent.com/python-visualization/folium/master/examples/data"
)
state_geo = f"{url}/us-states.json"
state_unemployment = "Identified Serial killers in the United States in 60 years (1960-2020) - Numbers of Killers in each State.csv"
state_data = pd.read_csv(state_unemployment)

folium.Choropleth(
    geo_data=state_geo,
    name="choropleth",
    data=state_data,
    columns=["State", "Num of killer"],
    key_on="feature.id",
    fill_color="Oranges",
    fill_opacity=0.7,
    line_opacity=.1,
    legend_name="Num of killer",
).add_to(m)

folium.LayerControl().add_to(m)

m
```
